```python
import pandas as pd
import math
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("probes.csv")
```

```python
df
```

```python
intensities = []

for intens in df[" intens"]:
    intensities.append(int(intens[0:3]))

df["n"] = intensities
df
```

```python
df["FSPL"] = 20+93.5-(df["n"]+93.5)
df
```

```python
df["distance"] = 10**((df["FSPL"]-90.45-20*math.log10(2.4))/20)
```

```python
df
```

```python
df["distance"].plot()
```

```python
df.describe()
```

```python
df["distance"].hist()
```

```python
real_distances = []

for distance in df["distance"]:
    if df["distance"].min() < distance <= 0.558842:
        real_distances.append(distance*3)
    elif 0.558842 < distance <= 1.115036:
        real_distances.append(distance*3)
    elif 1.115036 < distance <= 1.767212:
        real_distances.append(distance*4)
    else:
        real_distances.append(distance*4.3)


df["real_distances"] = real_distances

df
```

```python
df.describe()
```

```python
df["real_distances"].hist()
```
